```python
# Frequency Table of at_fault and statewide_vehicle type

fault_grp = dfp.groupby(["at_fault"])
```

```python
# Count and Percentage for At Fault Vehicle Types

fault_ct = fault_grp['statewide_vehicle_type'].value_counts().loc[1]

fault_per = fault_grp['statewide_vehicle_type'].value_counts(normalize=True).loc[1].mul(100).round(3).astype(str) + '%'

pd.DataFrame({'Count':fault_ct, 'Percentage':fault_per})
```
